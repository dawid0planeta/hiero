```python
import matplotlib.pyplot as plt

from skimage import data
from skimage.color import rgb2hsv

rgb_img = data.coffee()
hsv_img = rgb2hsv(rgb_img)
hue_img = hsv_img[:, :, 0]
value_img = hsv_img[:, :, 2]

fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(8, 2))

ax0.imshow(rgb_img)
ax0.set_title("RGB image")
ax0.axis('off')
ax1.imshow(hue_img, cmap='hsv')
ax1.set_title("Hue channel")
ax1.axis('off')
ax2.imshow(value_img)
ax2.set_title("Value channel")
ax2.axis('off')

fig.tight_layout()
```

```python
import skimage
from skimage.color import rgb2gray
from skimage.io import imread
from skimage.filters import threshold_mean
import numpy as np
import matplotlib.pyplot as plt

image = imread("./1.jpeg")
image_negatyw = np.invert(image)
image_grey = rgb2gray(image)
image_mean = threshold_mean(image_grey)
image_thresh = image_grey > image_mean
image_gray_negatyw = rgb2gray(image_negatyw)

fig, (ax0, ax1, ax2, ax3, ax4) = plt.subplots(ncols=5, figsize=(8, 2))

ax0.imshow(image)
ax0.axis("off")
ax1.imshow(image_grey, cmap=plt.cm.gray)
ax1.axis("off")
ax2.imshow(image_negatyw)
ax2.axis("off")
ax3.imshow(image_gray_negatyw, cmap=plt.cm.gray)
ax3.axis("off")
ax4.imshow(image_thresh, cmap=plt.cm.gray)
ax4.axis("off")

fig.tight_layout()
```

```python
import matplotlib.pyplot as plt
from skimage.filters import gaussian
from skimage.transform import swirl, warp

# Set the size of the figure
plt.rcParams["figure.figsize"] = [3.00, 3.00]

# Automatically adjust subplot parameters to give specified padding
plt.rcParams["figure.autolayout"] = True

# Loop through all the hieroglyphs
for i in range(0x13000, 0x13000 +2): #0x1342F+1):
    # Create a figure with the hieroglyph
    plt.text(0.35, 0.5, s=chr(i), fontsize=70, family="Noto Sans Egyptian Hieroglyphs")
    plt.axis("off")

    # Save the original image
    plt.savefig(f'baza/{i:X}_original.jpg')

    # Loop through 10 times to generate 10 distorted images
    for j in range(10):
        # Distort the image in different ways

        img = swirl(plt.imread(f'baza/{i:X}_original.jpg'), rotation=0, strength=j+1, radius=j+1)
        
        
        img = gaussian(img, sigma=j+1)
        print(img)

        # Save the distorted image
        plt.imsave(f'baza/{i:X}_{j}.jpg', img)
        plt.clf()
        
        
```
